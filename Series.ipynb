Series
- Series is an one-dimensional array with axis labels
- It supports both integer and label-based index but index must be of hashable type
- If we do not specify index it will assign integer zero-based index

Syntax: s = pd.Series(data,index,dtype,copy=False)

```python
import pandas as pd

s = pd.Series([1, 3, 5, 7, 9, 11])
print(s)
```

data - values stored such as list, tuple, etc
index - Defines the labels or index for the Series.
dtype - defines wha datatype the value will be
copy - defines whether the data is orignal

```python
import pandas as pd

s = pd.Series([1, 3, 5, 7, 9, 11])
print("S[0] = ", s[0])
b = s[0] + s[1]
print("Sum = ", b)
```

```python
import pandas as pd
s = pd.Series([1, 3, 5, 7, 9, 11], dtype='str')
print("S[0] = ", s[0])
b = s[0] + s[1]
print("Sum = ", b)
```
