Data frames are two dimensional data structure, i.e. data is aligned in a tabular format in rows  and columns.
Syntax: df = pd.DataFrame(data,index,columns,dtype,copy=False)

Grabbing the column
Syntax: print(df['collumn_name'])

Grabbing the row
Syntax: (using lables) - print(df.loc[index]) 
        (using zero based index) - print(df.iloc[index])

Grabbing single value
Syntax: print(df.loc[index, 'column name'])

Deleting row
Syntax:df.drop('index',inplace=True)
print(df)

Creating new column
Syntax: df['new_column'] = df['column_1'] + df['column_2'] + df['column_3'] + df['column_4']...
print(df)

Deleting column and row
df.drop('Column_name',axis=1,inplace=True)
print(df)

getting subset of dataframe
print(df.loc[[Rows], [[Columns]])

Conditional selection 
print(df[column] (condition))

Setting/Resetting 
df.index.name('RollNo')
df.set_index('PDS') 
df.reset_index()


```python
import numpy as np
import pandas as pd
randArr = np.random.randint(0,100,20).reshape(5,4)
df = pd.DataFrame(randArr,np.arange(101,106,1),['PDS','Algo','SE','INS'])
print(df)
print(df['PDS'])
print(df.loc[101]) 
Syntax: print(df.loc[101, 'PDS'])
df.drop('103',inplace=True)
print(df)
df['total'] = df['PDS'] + df['Algo'] + df['SE'] + df['INS']
print(df)
df.drop('total',axis=1,inplace=True)
print(df)
print(df.loc[[101,104], [['PDS','INS']])
print(df>50)
dfBool = df['PDS'] > 50
print(df[dfBool])
df.index.name('RollNo')
```
